```python
#TASK - 4
mylist = [10,30,20,70,11,15,22,16,58,100,12,56,70,80]
list1 = []
list2 = []
list3 = []

def bubble_sort (f_list):
  for x in range (0, len(f_list) - 1):
    for y in range (0, len(f_list) - x - 1):
      if f_list[y] > f_list[y + 1]:
        f_list[y], f_list[y + 1] = f_list[y+1], f_list[y]
  return f_list

for x in range (0, len(mylist)):
  if x % 2 == 0:
    list1.append(mylist[x])
  else:
    list2.append(mylist[x])
  
list1 = bubble_sort(list1)
list2 = bubble_sort(list2)[::-1]

for x in range (0, len(list1)):
  list3.append(list1[x])
  list3.append(list2[x])

print (list3)
```

```python
# TASK - 9
import math
points = [(1,7), (4,5), (-1,7), (-2,0), (1,1), (5,-1)]
dict1 = {}
lowest = None
coords = None

def formula (x, y):
  result = math.sqrt((x**2) + (y**2))
  return result

for x in points:
  dist = formula (x[0], x[1])
  dict1.update({x:dist})

for x in dict1:
    if lowest is None or dict1[x] < lowest:
        coords = x
        lowest = dict1[x]

print ("Minimum distance =", lowest)
print ("Here the closest point is", coords, "which has a distance of", lowest, "from the origin.")
```
