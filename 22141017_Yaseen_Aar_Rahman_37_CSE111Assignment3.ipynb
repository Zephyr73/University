```python
# TASK - 4

class Vehicle:

    def __init__ (self):
        self.coordinates = [0, 0]

    def print_position(self):
        print (tuple(self.coordinates))

    def moveUp(self):
        self.coordinates[1] = self.coordinates[1] + 1

    def moveDown(self):
        self.coordinates[1] = self.coordinates[1] - 1

    def moveLeft(self):
        self.coordinates[0] = self.coordinates[0] - 1

    def moveRight(self):
        self.coordinates[0] = self.coordinates[0] + 1


car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
# TASK - 5

class Shape:
    def __init__ (self, name, side1, side2):
        self.name = name
        self.side1 = side1
        self.side2 = side2

    def area(self):
        if self.name == 'Triangle':
            area = (self.side1 * self.side2)/2
        elif self.name == 'Square':
            area = self.side1 * self.side2
        elif self.name == 'Rhombus':
            area = (self.side1 * self.side2)/2
        elif self.name == 'Rectangle':
            area = self.side1 * self.side2
        else:
            area = 'Shape unknown'
        print (f'Area: {area}')

triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
# TASK - 6

class Calculator: 
    def __init__ (self):
        print ('Calculator is ready!')

    def calculate(self, num1, num2, symb):
        self.num1 = num1
        self.num2 = num2
        self.symb = symb

        if self.symb == '+':
            self.result = num1 + num2
        elif self.symb == '-':
            self.result = num1 - num2
        elif self.symb == '/':
            self.result = num1 / num2
        elif self.symb == '*':
            self.result = num1 * num2
        
        return self.result
    
    def showCalculation(self):
        print (f'{self.num1} {self.symb} {self.num2} = {self.result}')

c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
# TASK - 7

class Student:
    def __init__ (self, name, id, dept, results):
        self.name = name
        self.id = id
        self.dept = dept
        self.results = results


    def calculate_CGPA(self):
        total = 0
        for x in self.results:
            total = total + (x * 3)
        self.cgpa = total/(len(self.results) * 3)

    def print_details(self):
        print (f'Name: {self.name}, ID: {self.id}')
        print (f'Department {self.dept}')

        if self.cgpa > 3.80:
            print (f'CGPA: {self.cgpa}')
            print ("Your academic standing is 'Highest Distinction'")
        elif self.cgpa > 3.65 and self.cgpa <= 3.80:
            print (f'CGPA: {self.cgpa}')
            print ("Your academic standing is 'High Distinction'")
        elif self.cgpa > 3.5 and self.cgpa <= 3.65:
            print (f'CGPA: {self.cgpa}')
            print ("Your academic standing is 'Distinction'")
        elif self.cgpa > 2 and self.cgpa <= 3.5:
            print (f'CGPA: {self.cgpa}')
            print ("Your academic standing is 'Satisfactory'")
        elif self.cgpa <= 2:
            print (f'CGPA: {self.cgpa}')
            print ("Sorry, you can't graduate")


s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================") 
s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()
```

```python
# TASK - 8

class Shinobi:
    def __init__ (self, name, rank):
        self.name = name
        self.rank = rank
        self.mission = 0
        self.salary = 0
    
    def calSalary (self, mission):
        self.mission = mission
        if self.rank == 'Genin':
        
            self.salary = self.mission * 50
        elif self.rank == 'Chunin':
            self.salary = self.mission * 100
        else:
            self.salary = self.mission * 500

    def printInfo (self):
        print (f'Name: {self.name}')
        print (f'Rank: {self.rank}')
        print (f'Number of mission: {self.mission}')
        print (f'Salary: {self.salary}')

    def changeRank(self, new_rank):
        self.rank = new_rank

naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```

```python
# TASK - 9
class Programmer:
    def __init__ (self, name, lang, exp):
        self.name = name
        self.lang = lang
        self.exp = exp
        print ('Horray! A new programmer is born')

    def printDetails (self):
        print (f'Name: {self.name}')
        print (f'Language: {self.lang}')
        print (f'Experience: {self.exp} years.')

    def addExp (self, add):
        print ('Updating experience of Jon Snow')
        self.exp = self.exp + add

p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()  
```

```python
# TASK - 10

class UberEats:
    def __init__ (self, name, number, area):
        self.name = name
        self.number = number
        self.area = area
        print (f'{self.name}, welcome to UberEats!')
    
    def add_items (self, item1, item2, price1, price2):
        self.item1 = item1
        self.item2 = item2
        self.price1 = price1
        self.price2 = price2
        
        self.dict1 = {self.item1:self.price1, self.item2:self.price2}
    
    def print_order_detail (self):
        print (f'User details: {self.name}, Phone: {self.number}, Address: {self.area}')
        print (f'Orders: {self.dict1}')
        return f'Total Paid Amount: {self.dict1[self.item1] + self.dict1[self.item2]}'



order1 = UberEats("Shakib", "01719658xxx", "Mohakhali")
print("=========================")
order1.add_items("Burger", "Coca Cola", 220, 50)
print("=========================")
print(order1.print_order_detail())
print("=========================")
order2 = UberEats ("Siam", "01719659xxx", "Uttara")
print("=========================")
order2.add_items("Pineapple", "Dairy Milk", 80, 70)
print("=========================")
print(order2.print_order_detail())
```
