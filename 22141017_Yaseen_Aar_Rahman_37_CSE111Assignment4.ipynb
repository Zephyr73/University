```python
#TASK 1

class Customer:
    def __init__ (self, name = None):
        self.name = name

    def greet(self, name = None):
        if name is None:
            print ('Hello!')
        else:
            print ('Hello!', self.name)

    def purchase (self, *items):
        self.items = items
        print (f'{self.name}, you purchased {len(items)} item(s):')
        for x in items:
            print (x)

customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
# Task - 2

class Panda:
    def __init__ (self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age

    def sleep(self, hours = None):
        self.hours = hours
        if self.hours is None:
            self.food = 'bamboo leaves'
        elif self.hours >= 3 and self.hours <= 5:
            self.food = 'Mixed Veggies'
        elif self.hours >= 6 and self.hours <= 8:
            self.food = 'Eggplant & Tofu'
        elif self.hours >= 9 and self.hours <= 11:
            self.food = 'Broccoli Chicken'
        
        if self.hours is not None:
            return f'{self.name} sleeps {self.hours} hours daily and should have {self.food}'
        else:
            return f"{self.name}'s duration is unknown thus should have only {self.food}"

panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)

print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))

print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))

print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
 
```

```python
# Task - 4

class Student:
    def __init__ (self, name = None):
        self.name = name
    
    def quizcalc (self, *score):
        self.avg = 0
        for x in score:
            self.avg = self.avg + x
        self.avg = self.avg/3

    def printdetail(self):
        if self.name is None:
            print  ('Hello default student')
        else:
            print (f'Hello {self.name}')
        print (f'Your average quiz score is {self.avg}')

s1 = Student()
s1.quizcalc(10)
print('--------------------------------')
s1.printdetail()
s2 = Student('Harry')
s2.quizcalc(10,8)
print('--------------------------------')
s2.printdetail()
s3 = Student('Hermione')
s3.quizcalc(10,9,10)
print('--------------------------------')
s3.printdetail() 
```

```python
# Task - 5

class Student:
    def __init__ (self, name, id, dept = 'CSE'):
        self.name = name
        self.id = id
        self.dept = dept
    def dailyEffort (self, hours):
        self.hours = hours
        if self.hours <= 2:
            self.suggestion = 'Should give more effort!'
        elif self.hours <= 4 and self.hours > 2:
            self.suggestion = 'Keep up the good work!'
        elif self.hours > 4:
            self.suggestion = 'Excellent! Now motivate others'

    def printDetails(self):
        print (f'Name: {self.name}')
        print (f'ID: {self.id}')
        print (f'Department: {self.dept}')
        print (f'Daily Effort: {self.hours} hour(s)')
        print (f'Suggestion: {self.suggestion}')

harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()	
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
# Task - 6

class Patient:
    def __init__ (self, name, age):
        self.name = name
        self.age = age
    
    def add_Symptom(self, *symptom):
        self.symptom = symptom

    def printPatientDetail(self):
        symptoms = ''
        for x in range(0, len(self.symptom)):
            if x!= len(self.symptom) - 1:
                symptoms = symptoms + self.symptom[x] +', '
            else:
                symptoms = symptoms + self.symptom[x]

        print (f'Name: {self.name}')
        print (f'Age: {self.age}')
        print (f'Symptoms: {symptoms}')

p1 = Patient('Thomas', 23)
p1.add_Symptom('Headache')
p2 = Patient('Carol',  20)
p2.add_Symptom('Vomiting', 'Coughing')
p3 = Patient('Mike', 25)
p3.add_Symptom('Fever', 'Headache', 'Coughing')
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
```

```python
# Task - 7

class Match:
    def __init__ (self, team):
        print ('5..4..3..2..1..Play!!')
        team = team.split('-')
        self.batting_team = team[0]
        self.bowling_team = team[1]
        self.run = 0
        self.over = 0
        self.wicket = 0
    
    def add_run (self, run):
        self.run = self.run + run

    def add_over (self, over):
        self.over = self.over + over
        if self.over > 5:
            print (f'Warning! Cannot add {over} over/s (5 over match)')
            self.over = self.over - over

    def add_wicket (self, wicket):
        self.wicket = self.wicket + wicket
    
    def print_scoreboard(self):
        print (f'Batting Team: {self.batting_team}')
        print (f'Bowling Team: {self.bowling_team}')
        return f'Total Runs: {self.run} Wickets: {self.wicket} Overs: {self.over}'

match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(5)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())
```

```python
# Task - 8

class ParcelKoro:
    def __init__(self, name = 'No name set', product_weight = 0):
        self.name = name
        self.product_weight = product_weight
    
    def calculateFee (self, location = None):
        if self.product_weight == 0:
            self.total_fee = 0
        elif location is None:
            self.total_fee = (self.product_weight * 20) + 50
        else:
            self.total_fee = (self.product_weight * 20) + 100

    def printDetails(self):
        print (f'Customer Name: {self.name}')
        print (f'Product Weight: {self.product_weight}')
        print (f'Total fee: {self.total_fee}')


print("**********************")
p1 = ParcelKoro()
p1.calculateFee()
p1.printDetails()
print("**********************")
p2 = ParcelKoro('Bob The Builder')
p2.calculateFee()
p2.printDetails()
print("----------------------------")
p2.product_weight = 15
p2.calculateFee()
p2.printDetails()
print("**********************")
p3 = ParcelKoro('Dora The Explorer', 10)
p3.calculateFee('Dhanmondi')
p3.printDetails()
```

```python
# Task - 9

class Batsman:
    def __init__(self, val1, val2, val3 = None):
        if type(val1) == int:
            self.name = 'New Batsman'
            self.runs_scored = val1
            self.balls_faced = val2
        elif type(val1) == str:
            self.name = val1
            self.runs_scored = val2
            self.balls_faced = val3

    def printCareerStatistics(self):
        print (f'Name: {self.name}')
        print (f'Runs Scored: {self.runs_scored} Balls Faced: {self.balls_faced}')

    def battingStrikeRate(self):
        rate = self.runs_scored / self.balls_faced * 100
        return rate

    def setName(self, name):
        self.name = name


b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
# Task - 12

class TaxiLagbe:
    def __init__ (self, taxi_num, city):
        self.taxi_num = taxi_num
        self.city = city
        self.name_list = []
        self.price_list = []

    def addPassenger (self, *list1):
        temp_list = []
        for x in list1:
            name = x.split('_')
            self.name_list.append(name[0])
            temp_list.append(name[0])
            self.price_list.append(int(name[1]))
        
        if len(self.name_list) > 4:
            self.name_list.pop()
            self.price_list.pop()
            print ('Taxi Full! No more passengers can be added.')
        else:
            for x in temp_list:
                print (f'Dear {x}! Welcome to TaxiLagbe')
    
    def printDetails (self):
        total = 0
        for x in self.price_list:
            total = total + x

        print (f'Trip info for Taxi number: {self.taxi_num}')
        print (f'This taxi can cover only {self.city} area')
        print (f'Total passengers: {len(self.name_list)}')
        print ('Passenger lists:')
        for x in range (0, len(self.name_list)):
            if x != len(self.name_list) - 1:
                print (self.name_list[x] + ', ', end ='')
            else:
                print (self.name_list[x])
        print (f'Total collected fare: {total} Taka')
        

taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
# Task - 13

class Account:
    def __init__ (self, name = 'Default Account', balance = 0.0):
        self.name = name
        self.balance = balance

    def details(self):
        print (self.name)
        return float(self.balance)

    def withdraw(self, amount):
        minimum_balance = 3070
        if self.balance == 0 or self.balance < amount or (self.balance - amount) <= minimum_balance:
            print ('Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.')
        else:
            self.balance = self.balance - amount
            print ('Withdraw successful! New balance is:')
            print (float(self.balance))

a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```

```python
# Task - 14

class StudentDatabase:
    def __init__ (self, name, id):
        self.name = name
        self.id = id
        self.grades = {}

    def calculateGPA(self, list1, session):
        courses = []
        cgpa = []
        temp_dict = {}
        avg_cgpa = 0
        for x in list1:
            courses.append(x.split(': ')[0])
            cgpa.append(x.split(': ')[1])

        for x in cgpa:
            avg_cgpa = avg_cgpa + float(x)

        avg_cgpa = round(avg_cgpa/len(cgpa), 2)
        temp_dict[tuple(courses)] = avg_cgpa
        self.grades[session] = temp_dict

    def printDetails(self):
        print (f'Name: {self.name}')
        print (f'ID: {self.id}')
        for session, dict1 in self.grades.items():
            print (f'Courses taken in {session}')
            for courses, cgpa in dict1.items():
                for course in courses:
                    print (course)
                print(cgpa)


s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
