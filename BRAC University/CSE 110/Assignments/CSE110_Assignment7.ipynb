```python
#--------------------[TASK - 7]--------------------#

def bubble_sort(f_list):
    for x in range(0, len(f_list) - 1):
        for y in range(0, len(f_list) - x - 1):
            if f_list[y] > f_list[y + 1]:
                f_list[y], f_list[y + 1] = f_list[y+1], f_list[y]
    return f_list


def str_to_int(f_list):
    for x in range(len(f_list)):
        f_list[x] = int(f_list[x])
    return f_list


usr_input = str(input("Enter a list: "))
usr_input2 = str(input("Enter a list: "))
usr_input = usr_input[1:-1].replace(' ', '').split(',')
usr_input2 = usr_input2[1:-1].replace(' ', '').split(',')

final_list = bubble_sort(str_to_int(usr_input + usr_input2))

if len(final_list) % 2 == 0:
    print((final_list[len(final_list)//2] + final_list[(len(final_list) - 1) // 2])/2)
else:
    print(final_list[len(final_list)//2])
```

```python
#--------------------[TASK - 8]--------------------#

list1 = [1, -8, 4, -7, -20, 26, 70, -85]
list2 = []
dict1 = {}
dict2 = {}


def bubble_sort(f_list):
    for x in range(0, len(f_list) - 1):
        for y in range(0, len(f_list) - x - 1):
            if f_list[y] > f_list[y + 1]:
                f_list[y], f_list[y + 1] = f_list[y + 1], f_list[y]
    return f_list


list1 = bubble_sort(list1)

for x in range(0, len(list1)):
    for y in range(x + 1, len(list1)):
        dict1.update({list1[x]+list1[y]: [list1[x], list1[y]]})

for x in dict1:
    if x < 0:
        dict2.update({-x: dict1[x]})
    else:
        dict2.update({x: dict1[x]})

for x in dict2:
    list2.append(x)
list2 = bubble_sort(list2)
lowest = list2[0]

print("Two pairs which have the smallest sum =", dict2[lowest][1], "and", dict2[lowest][0])
```

```python
#--------------------[TASK - 9]--------------------#

import math
points = [(5, 3), (2, 9), (-2, 7), (-3, -4), (0, 6), (7, -2)]
list1 = []
dict1 = {}
coords = None


def bubble_sort(f_list):
    for x in range(0, len(f_list) - 1):
        for y in range(0, len(f_list) - x - 1):
            if f_list[y] > f_list[y + 1]:
                f_list[y], f_list[y + 1] = f_list[y+1], f_list[y]
    return f_list


def formula(x, y):
    result = result = math.sqrt((x**2) + (y**2))
    return result


for x in points:
    dist = formula(x[0], x[1])
    dict1.update({x: dist})

for x in dict1:
    list1.append(dict1[x])
list1 = bubble_sort(list1)

for key, val in dict1.items():
    if val == list1[0]:
        coords = key

print("Minimum distance =", list1[0])
print("Here the closest point is", coords, "which has a distance of", list1[0], "from the origin.")
```
