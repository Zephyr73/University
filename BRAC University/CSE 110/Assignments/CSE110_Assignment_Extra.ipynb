```python
def check_awesome(list1):
    for x in range (0, len(list1)):
        list1[x] = str(list1[x])

    for elem in list1:
        prev = elem[0]
        if len(elem) == 1:
            print ("False")
            break
        else:
            for y in range (1, len(elem)):
                if int(elem[y]) >= int(prev):
                    flag = False
                    break
                else:
                    prev = elem[y]
                    flag = True
        if flag == True:
            print (elem, "is an awesome number")
        if flag == False:
            print (elem, "is a not-so-awesome number")

check_awesome([976321, 321, 9763221, 9742, 876, 3211])
```

```python
def find_max_min(list1):
    highest = None
    lowest = None
    tuple1 = []
    for x in list1:
        if highest is None or x > highest:
            highest = x
        if lowest is None or x < lowest:
            lowest = x
    tuple1.append(highest)
    tuple1.append(lowest)
    tuple1 = tuple(tuple1)
    h, l = tuple1

    print ("Returned value from find_max_min() is:", tuple1)
    print ("Number with maximum value is", h)
    print("Number with minimum value is", l)

find_max_min ([976321, 321, 9763221, 9742, 876, 3211, 976321, 9742])
```

```python
def unique_number(tuple1):
    new_number1 = 0
    new_number2 = 1
    result1 = ""
    result2 = ""

    for x in tuple1:
        if x % 2 == 0:
            new_number1 = new_number1 + x
        if x % 2 != 0:
            new_number2 = new_number2 * x
    
    for x in range(0, len(str(new_number1))):
        if x % 2 == 0:
            result1 = result1 + str(new_number1)[x]
    for x in range(0, len(str(new_number2))):
        if x % 2 != 0:
            result2 = result2 + str(new_number2)[x]
    
    print ("The unique number is:", result1 + result2)

usr_input = str(input("Enter a list of numbers: "))
usr_input = usr_input[1:-1].replace(' ', '').split(',')

for x in range (0, len(usr_input)):
    usr_input[x] = int(usr_input[x])

unique_number(usr_input)
```
