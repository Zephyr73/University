```python
# TASK - 1

class Student:
    ID = 0

    def __init__ (self, name, dept, age, cgpa):
        Student.ID = Student.ID + 1
        self.name = name
        self.dept = dept
        self.age = age
        self.cgpa = cgpa

    @classmethod
    def from_String(cls, string):
        list1 = string.split('-')
        cls.name = list1[0]
        cls.dept = list1[1]
        cls.age = list1[2]
        cls.cgpa = list1[3]
        return cls(cls.name, cls.dept, cls.age, cls.cgpa)
    
    def get_details(self):
        print (f'ID: {Student.ID}')
        print (f'Name: {self.name}')
        print (f'Department: {self.dept}')
        print (f'Age: {self.age}')
        print (f'CGPA: {self.cgpa}')

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details() 
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details() 
```

```python
class Assassin:
    total = 0

    def __init__ (self, name, rate):
        self.name = name
        self.rate = rate
        Assassin.total = Assassin.total + 1

    def printDetails (self):
        print (f'Name: {self.name}')
        print (f'Success rate: {self.rate}')
        print (f'Total number of Assassin: {Assassin.total}')

    @classmethod
    def failureRate (cls, name, rate):
        cls.name = name
        cls.rate = 100 - rate
        return Assassin(cls.name, cls.rate)

    @classmethod
    def failurePercentage (cls, name, rate):
        cls.name = name
        cls.rate = 100 - int(rate[:-1])
        return Assassin(cls.name, cls.rate)


john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
# TASK - 5

from datetime import date
class Employee:
    def __init__ (self, name, workingPeriod):
        self.name = name
        self.workingPeriod = workingPeriod

    @classmethod
    def employeeByJoiningYear (cls, name, year):
        cls.workingPeriod = date.today().year - year
        return cls(name, cls.workingPeriod)

    def workingPeriod (self):
        return self.workingPeriod

    @staticmethod
    def experienceCheck (year, gender):
        if year < 3:
            if gender == 'male':
                return 'He is not experienced'
            elif gender == 'female':
                return 'She is not experienced'
        else:
            if gender == 'male':
                return 'He is experienced'
            elif gender == 'female':
                return 'She is experienced'

employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
# TASK - 7

class Cat:
    Number_of_cats = 0
    color = 'White'
    action = 'sitting'

    def __init__(self, color, action):
        self.color = color
        self.action = action
        Cat.Number_of_cats = Cat.Number_of_cats + 1

    @classmethod
    def no_parameter(cls):
        return Cat(cls.color, cls.action)

    @classmethod
    def first_parameter(cls, color):
        cls.color = color
        cls.action = Cat.action
        return Cat(cls.color, cls.action)

    @classmethod
    def second_parameter(cls, action):
        cls.color = 'Grey'
        cls.action = action
        return Cat(cls.color, cls.action)

    def printCat(self):
        print(f'{self.color} cat is {self.action}')

    def changeColor(self, color):
        self.color = color

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
# TASK - 8

import math

class Cylinder:
    radius = 5
    height = 18
    
    def __init__(self, radius, height):
        print(f'Default radius={Cylinder.radius} and height={Cylinder.height}')
        Cylinder.radius = radius
        Cylinder.height = height
        print(f'Updated: radius={Cylinder.radius} and height={Cylinder.height}')

    @staticmethod
    def area(radius, height):
        print(f'Area: {2*math.pi*(radius)**2 + 2*math.pi*radius*height}')
    
    @staticmethod
    def volume(radius, height):
        print(f'Volume: {math.pi*height*(radius)**2}')
    
    @classmethod
    def swap(cls, height, radius):
        return cls(radius, height)
    
    @classmethod
    def changeFormat(cls, rad_height):
        radius = rad_height.split('-')[0]
        height = rad_height.split('-')[1]
        return cls(float(radius), float(height))
    


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
