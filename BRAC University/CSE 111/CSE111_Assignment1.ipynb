# STRING

```python
# TASK - 1

lower = 0
upper = 0
usr_input = str(input('Enter a text: '))

for x in usr_input:
    if ord(x) >= 65 and ord(x) <= 90:
        upper = upper + 1
    else:
        lower = lower + 1

if upper > lower:
    print(usr_input.upper())
else:
    print(usr_input.lower())
```

```python
# TASK - 2

letter = 0
number = 0
usr_input = str(input('Enter a text: '))

for x in usr_input:
    if (ord(x) >= 65 and ord(x) <= 90) or (ord(x) >= 97 and ord(x) <= 122):
        letter = letter + 1
    else:
        number = number + 1

if number == 0 and letter > 0:
    print('WORD')
elif number > 0 and letter == 0:
    print('NUMBER')
elif number > 0 and letter > 0:
    print('MIXED')
```

```python
# TASK - 3

text = str(input("Enter a text: "))
caps_idx = []

for x in range(0, len(text)):
    if ord(text[x]) >= 65 and ord(text[x]) <= 90:
        caps_idx.append(x)
if caps_idx[1] - caps_idx[0] == 1:
    print("BLANK")
else:
    print(text[caps_idx[0] + 1: caps_idx[1]])
```

```python
# TASK - 4

str1 = str(input("Enter first word: "))
str2 = str(input("Enter second word: "))
str3 = ''

for x in str1:
    if x in str2:
        str3 = str3 + x
for x in str2:
    if x in str1:
        str3 = str3 + x
if str3 == '':
    print('Nothing in common.')
else:
    print(str3)
```

```python
# TASK - 5

number = 0
upper_char = 0
lower_char = 0
sp_character = 0

password = str(input("Enter the password: "))
sp_char = ['_', '$', '#', '@']

for x in password:
    if (ord(x) >= 65 and ord(x) <= 90):
        upper_char = upper_char + 1
    elif (ord(x) >= 97 and ord(x) <= 122):
        lower_char = lower_char + 1
    elif (ord(x) >= 48 and ord(x) <= 57):
        number = number + 1
    elif x in sp_char:
        sp_character = sp_character + 1

if upper_char > 0 and number > 0 and sp_character > 0 and lower_char > 0:
    print('OK')
if sp_character == 0:
    print('Special Character missing')
if number == 0:
    print('Digit missing')
if lower_char == 0:
    print('Lowercase character missing')
if upper_char == 0:
    print('Uppercase character missing')
```

# LIST

```python
# TASK - 1

list1 = []
list2 = []

while True:
    usr_input = input('Enter a number: ')
    if usr_input == 'STOP':
        break
    else:
        list1.append(usr_input)

for x in list1:
    if x not in list2:
        list2.append(int(x))
        list2.append(1)
    elif int(x) in list2:
        list2[list2.index(x) + 1] += 1

for x in range(0, len(list2), 2):
    print(list2[x], '-', list2[x + 1], 'times')
    
```

```python
# TASK - 2

list1 = []
highest = None
ans_list = None
total = 0

while True:
    usr_input = str(input("Enter a List separated by space: "))
    if usr_input == 'stop':
        break
    else:
        usr_input = usr_input.split(" ")
        for x in range (0, len(usr_input)):
            usr_input[x] = int(usr_input[x])
        list1.append(usr_input)

for x in list1:
    for y in x:
        total = total + y
    if highest is None or total > highest:
        highest = total
        ans_list = x
        total = 0

print (highest)
print (ans_list)
```

```python
# TASK - 3

list1 = []

def jumper_checker(f_list):
    list2 = []
    limit = len(f_list) - 1

    for x in range(0, len(f_list)):
        difference = None
        if x != len(f_list) - 1:
            difference = f_list[x] - f_list[x + 1]
            if difference < 0:
                difference = difference * - 1
            list2.append(difference)
        
    for x in list2:
        if x <= limit:
            flag = True
        else:
            flag = False
    if flag == True:
        print("UB Jumper")
    if flag == False:
        print("Not UB jumper")

while True:
    usr_input = str(input("Enter a List separated by space: "))
    if usr_input == 'stop':
        break
    else:
        list1 = usr_input.split(" ")
        for x in range(0, len(list1)):
            list1[x] = int(list1[x])
        jumper_checker (list1)
```

```python
# TASK - 1

list1 = []
dict1 = {'a': 100, 'b': 100, 'c' : 200, 'd': 300}
dict2 = {'a': 300, 'b': 200, 'd': 400, 'e': 200}

for key in dict2.keys():
    if key in dict1:
        dict1[key] = dict1[key] + dict2[key]
    else:
        dict1[key] = dict2[key]

for x in dict1.values():
    if x not in list1:
        list1.append(x)
list1.sort()

print(tuple(list1))
```

```python
# TASK - 2

dict1 = {}

while True:
    usr_input = str(input("Enter a number: "))
    if usr_input == 'stop':
        break
    else:
        if usr_input not in dict1.keys():
            dict1.update ({usr_input : 1})
        elif usr_input in dict1.keys():
            dict1[usr_input] = dict1[usr_input] + 1

for k, v in dict1.items():
    print (k, '-', v, 'times')
```

```python
# TASK - 1

def bmi_calculator(height, weight):
    height = height/100
    bmi = weight/(height**2)
    if bmi < 18.5:
        print ('Score is', str(round(bmi, 1)) + '.', 'You are Underweight')
    elif bmi >= 18.5 and bmi <= 24.9:
        print ('Score is', str(round(bmi, 1)) + '.', 'You are Normal')
    elif bmi >= 25 and bmi <= 30:
        print ('Score is', str(round(bmi, 1)) + '.', 'You are Overweight')
    else:
        print ('Score is', str(round(bmi, 1)) + '.', 'You are Obese')

bmi_calculator()
```

```python
# TASK - 2

def summation(minimum, maximum, div):
    total = 0
    for x in range (minimum, maximum):
        if x%div == 0:
            total = total + x
    print (total)

summation()
```

```python
# TASK - 5

def palindrome(text):
    text = text.replace(' ', '')
    if text == text[::-1]:
        print ('Palindrome')
    else:
        print ('Not a palindrome')

palindrome ('hello')
```
