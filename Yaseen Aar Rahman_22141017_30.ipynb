```python
digits = int(input("Enter a value: "))
last_sum = 0
first_sum = 0

if digits <=1:
    print ("Invalid Input")
else:
    for x in range (digits):
        num = int(input("Enter a number: "))
        last_sum = last_sum + (num % 10)
        first_sum = first_sum + (num // 10**(digits - 1))
    print ("Summation of every first digit:", first_sum)
    print ("Summation of every last digit:", last_sum)
```
